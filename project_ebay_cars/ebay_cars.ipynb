# Analyzing Used Car Listings on eBay Kleinanzeigen

In this guided project, we'll work with a dataset of used cars from eBay Kleinanzeigen, a classifieds section of the German eBay website.

## Importing Data

```python
import numpy as np
import pandas as pd
```

```python
# Reading csv file
autos = pd.read_csv('autos.csv', encoding='latin1')

# Informations about the table
autos.info()

# Displaying the first 5 rows of the table
autos.head()
```

## Data Cleaning

```python
# Cleaning column names
autos.columns = ['date_crawled', 'name', 'seller', 'offer_type', 'price', 'ab_test',
                 'vehicle_type', 'registration_year', 'gearbox', 'power_ps', 'model',
                 'odometer', 'registration_month', 'fuel_type', 'brand',
                 'unrepaired_damage', 'ad_created', 'num_photos', 'postal_code',
                 'last_seen']
autos.head()
```

```python
autos.describe(include='all')
```

```python
# Converting Price column data type
autos['price'] = autos['price'].str.replace('$', '')\
                               .str.replace(',', '')\
                               .astype(int)
```

```python
autos['price'].head()
```

```python
# Converting Odometer column data type
autos['odometer'] = autos['odometer'].str.replace('km', '')\
                                     .str.replace(',', '')\
                                     .astype(int)
```

```python
autos.rename({'odometer':'odometer_km'}, axis=1, inplace=True)
autos['odometer_km'].head()
```

```python
autos['odometer_km'].unique().shape
autos['odometer_km'].describe()
autos['odometer_km'].value_counts().sort_index(ascending=False).head(20)
```

```python
autos['price'].unique().shape
autos['price'].describe()
autos['price'].value_counts().sort_index(ascending=False).head(20)
```

```python
autos['price'].value_counts().sort_index().head(20)
```

```python
autos = autos[autos['price'].between(1,351000)]
autos['price'].describe()
```

```python
autos['date_crawled'].str[:10].value_counts(normalize=True, dropna=False).sort_index()
```

```python
autos = autos[autos['registration_year'].between(1900,2016)]
```

```python
autos.describe()
```

```python
autos['registration_year'].value_counts(normalize=True).head(10)
```

```python
brand_counts = autos['brand'].value_counts(normalize=True)
common_brands = brand_counts[brand_counts > 0.05].index
print(common_brands)
```

```python
brand_mean_prices = {}

for brand in common_brands:
    my_brand = autos[autos['brand'] == brand]
    mean_brand = my_brand['price'].mean()
    brand_mean_prices[brand] = int(mean_brand)

brand_mean_prices
```

```python
mean_mileage = {}

for brand in common_brands:
    my_brand = autos[autos['brand'] == brand]
    mileage_mean = my_brand['odometer_km'].mean()
    mean_mileage[brand] = int(mileage_mean)

mean_mileage
```

```python
mean_mileage = pd.Series(mean_mileage)
mean_price = pd.Series(brand_mean_prices)
mean_mileage
```

```python
brand_info = pd.DataFrame(mean_mileage, columns=['mean_mileage'])
brand_info
```
